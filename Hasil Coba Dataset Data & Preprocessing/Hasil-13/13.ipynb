```python
# 1. Transformasi Data - Binning (bins.csv)
```

```python
# Import Library

import pandas as pd
```

```python
# Membaca Dataset

df = pd.read_csv("bins.csv")
df.head()
```

```python
# Melakukan Binning

bins = [0, 10000, 50000, 100000]

labels = ["Murah", "Sedang", "Mahal"]

df["Kategori_Harga"] = pd.cut(
    df["Harga"],
    bins=bins,
    labels=labels
)
```

```python
# Hasil Transformasi

df
```

```python
# Melihat Jumlah Tiap Kategori

df["Kategori_Harga"].value_counts()
```

```python
# 2. Transformasi Data - Min-Max Normalization (diabetes.csv)
```

```python
# Import Library

import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
# Membaca Dataset

df = pd.read_csv("diabetes.csv")
df.head()
```

```python
# Informasi Dataset

df.info()
```

```python
# Melakukan Normalisasi

scaler = MinMaxScaler()

df_normalized = pd.DataFrame(
    scaler.fit_transform(df),
    columns=df.columns
)
```

```python
# Hasil Normalisasi

df_normalized.head()
```

```python
# Statistik Setelah Normalisasi

df_normalized.describe()
```
