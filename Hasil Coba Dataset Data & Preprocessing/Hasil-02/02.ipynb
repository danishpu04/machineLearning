```python
# import library

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load dataset

df = pd.read_csv('automobileEDA.csv')
df.head()
```

```python
# info dataset

df.info()
df.describe()
```

```python
# Menggunakan NumPy (Perhitungan)

print("Rata-rata harga:", np.mean(df['price']))
print("Harga tertinggi:", np.max(df['price']))
print("Harga terendah:", np.min(df['price']))
```

```python
# VISUALISASI DATA

# Histogram Harga (Matplotlib)
plt.figure()
plt.hist(df['price'])
plt.title('Distribusi Harga Mobil')
plt.xlabel('Price')
plt.ylabel('Jumlah')
plt.show()
```

```python
# Scatter Plot (Seaborn)

plt.figure()
sns.scatterplot(x='horsepower', y='price', data=df)
plt.title('Horsepower vs Price')
plt.show()
```

```python
# Bar Chart Rata-rata Harga per Body Style (Pandas + Matplotlib)

avg_price = df.groupby('body-style')['price'].mean()

plt.figure()
avg_price.plot(kind='bar')
plt.title('Rata-rata Harga Berdasarkan Body Style')
plt.xlabel('Body Style')
plt.ylabel('Average Price')
plt.show()
```

```python
# Boxplot (Seaborn)

plt.figure()
sns.boxplot(x='drive-wheels', y='price', data=df)
plt.title('Distribusi Harga Berdasarkan Drive Wheels')
plt.show()
```

```python
# Heatmap Korelasi (Seaborn + NumPy)

plt.figure()
corr = df.select_dtypes(include=[np.number]).corr()

sns.heatmap(corr, annot=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Scatter MPG vs Price

plt.figure()
sns.scatterplot(x='city-mpg', y='price', data=df)
plt.title('City MPG vs Price')
plt.show()
```
