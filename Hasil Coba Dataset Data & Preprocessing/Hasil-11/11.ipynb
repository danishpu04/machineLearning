```python
import pandas as pd
import numpy as np
```

```python
# Membaca Dataset

df = pd.read_csv('creditApprovalUCI.csv')
df.head()
```

```python
# Melihat Informasi Dataset

df.info()
```

```python
# Melihat Tipe Data

df.dtypes
```

```python
# Menampilkan Kolom Kategorikal

categorical_columns = df.select_dtypes(include='object').columns
print(categorical_columns)
```

```python
# Melakukan One Hot Encoding

df_encoded = pd.get_dummies(df, columns=categorical_columns)
```

```python
# Melihat Hasil Encoding

df_encoded.head()
```

```python
# Melihat Jumlah Kolom Setelah Encoding

print("Jumlah kolom sebelum encoding :", df.shape[1])
print("Jumlah kolom sesudah encoding :", df_encoded.shape[1])
```
