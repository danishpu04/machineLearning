```python
# Import Library

import pandas as pd
import numpy as np
```

```python
# Membaca Dataset

df = pd.read_csv('olympics.csv', header=1)
df.head()
```

```python
# Melihat Struktur Data

df.info()
df.describe()
```

```python
# Melihat Nama Kolom

df.columns
```

```python
# Mengganti Nama Kolom Pertama

df.rename(columns={df.columns[0]: 'Country'}, inplace=True)
```

```python
# Membersihkan Nama Kolom

df.columns = (
    df.columns
      .str.replace('?', '', regex=False)
      .str.replace('!', '', regex=False)
      .str.strip()
)

df.columns
```
