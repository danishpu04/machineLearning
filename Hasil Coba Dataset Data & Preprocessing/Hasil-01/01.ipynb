```python
# Import library

import pandas as pd
```

```python
# Load dataset

df = pd.read_csv('epl-goalScorer(20-21).csv')
df.head()
```

```python
# Memahami Struktur Data

df.info()
```

```python
# Memahami Struktur Data

df.describe()
```

```python
# Memahami Struktur Data

df.columns
```

```python
# Cek Data Kotor (Data Cleaning)

# Cek missing value

df.isnull().sum()
```

```python
# Cek duplikasi

df.duplicated().sum()
```

```python
# Seleksi Data

# Ambil kolom penting

df[['player_name', 'team_title', 'goals', 'assists']]
```

```python
# Analisis Sederhana

# Top scorer (gol terbanyak)

df.sort_values(by='goals', ascending=False).head(10)
```

```python
# Pemain dengan assist terbanyak

df.sort_values(by='assists', ascending=False).head(10)
```

```python
# Rata-rata gol

df['goals'].mean()
```

```python
# Total gol seluruh pemain

df['goals'].sum()
```

```python
# Analisis Lanjutan (Menelaah Data)

# Efisiensi pemain (goals vs xG)

df['efisiensi'] = df['goals'] - df['xG']
df[['player_name', 'goals', 'xG', 'efisiensi']].sort_values(by='efisiensi', ascending=False).head(10)

# Interpretasi:
# Nilai tinggi = pemain lebih tajam dari ekspektasi
# Nilai negatif = finishing kurang bagus
```

```python
# Kontribusi total (goal + assist)

df['kontribusi'] = df['goals'] + df['assists']
df[['player_name', 'kontribusi']].sort_values(by='kontribusi', ascending=False).head(10)
```

```python
# Pemain paling aktif (shots)

df.sort_values(by='shots', ascending=False).head(10)
```

```python
# Analisis Berdasarkan Tim

# Rata-rata gol per tim

df.groupby('team_title')['goals'].mean().sort_values(ascending=False)
```

```python
# Total gol per tim

df.groupby('team_title')['goals'].sum().sort_values(ascending=False)
```

```python
# Filtering Data

# Pemain dengan gol > 15

df[df['goals'] > 15]
```

```python
# Pemain posisi Forward (F)

df[df['position'].str.contains('F')]
```
